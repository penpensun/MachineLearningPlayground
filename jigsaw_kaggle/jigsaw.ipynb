```python
import pandas as pd;

short_train_csv = '/home/peng/Workspace/data/kaggle/jigsaw/short_train.csv';
short_train = pd.read_csv(short_train_csv);

def classify_target (x):
    if x == 0:
        return 0;
    if x > 0 and x<= 0.2:
        return 1;
    if x > 0.2 and x <= 0.4:
        return 2;
    if x >0.4 and x <= 0.6:
        return 3;
    if x >0.6 and x <= 0.8:
        return 4;
    if x> 0.8 and x <= 1.0:
        return 5;

short_train['target_class'] = short_train['target'].map(classify_target)
short_train
```
